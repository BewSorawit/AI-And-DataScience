# Linestyle

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3, 8, 1, 10])
# plt.plot(ypoints, linestyle='dotted')
plt.plot(ypoints, ls='--',color='r')
plt.show()
```

### Line Width

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, linewidth='10')
plt.show()
```

### Multiple Lines

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, c='r')
plt.plot(x2, y2, ls='--', c='#0000FF')
plt.show()
```
