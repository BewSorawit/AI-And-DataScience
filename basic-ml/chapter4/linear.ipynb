### Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = 2*x+1
```

```python
print(x[:10])
```

```python
print(y[:10])
```

#### plot

```python
plt.plot(x,y,'r',label = 'y=2x+1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="upper left")
plt.title("Graph y = 2x+1")
plt.grid()
plt.show()
```

#### scatter

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="upper left")
plt.title("Graph y = 2x+1")
plt.grid()
plt.show()
```
