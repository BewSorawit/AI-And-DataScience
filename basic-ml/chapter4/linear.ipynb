### Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
y = (2 * x) + 1
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.title('Linear Regression')
plt.grid()
plt.show()
```
